```python
# 英文を読み込んで単語リストを作る

def make_wordlist(file):
    word_lst = []                           # 単語リストを用意
    file_data = open(file)                  # ファイルをオープン
    for line in file_data:                  # 段落を取り出す
        tmp_lst = line.split()              # ①単語に分割してリストにする(改行は除く)

        for word in tmp_lst:                # リストから単語を取り出す
            word = word.rstrip('.,:!?)’”')# ②末尾のピリオド等を取り除く
            word = word.lstrip('(‘“')     # ③冒頭のクォートを取り除く
            word_lst.append(word)           # ④単語リストに追加する

    return word_lst
```

```python
file_name = 'alice.txt'
make_wordlist(file_name)     # ファイルを指定して単語リストを取得
```

```python
# 英文を読み込んで重複なしの単語リストを作る関数

def make_wordlist2(file):
    word_lst = []                       # 単語リストを用意
    file_data = open(file)              # ファイルをオープン
    for line in file_data:              # 段落を取り出す
      tmp_lst = line.split()            # 単語に分割してリストにする(改行は除かれる)

      for word in tmp_lst:              # リストから単語を取り出す
        word = word.rstrip('.,:!?)’”')# 末尾のピリオド等を取り除く
        word = word.lstrip('(‘“')     # 冒頭のクォートを取り除く
        if not word in word_lst:        # ①単語がすでに登録済みでないかチェック
            word_lst.append(word)       # 単語リストに追加する

    return word_lst
```

```python
file_name = 'alice.txt'
make_wordlist2(file_name)     # ファイルを指定して単語リストを取得
```
