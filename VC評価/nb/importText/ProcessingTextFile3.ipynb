```python
# 単語の出現回数をカウントして頻度表を作る関数

def get_frequency(file):
    freq = {}                           # 辞書を用意

    file_data = open(file)              # ファイルをオープン
    for line in file_data:              # ①段落を取り出す
        words = line.split()              # 単語に分割してリストにする(改行は除かれる)

        for word in words:                  # ②リストから単語を取り出す
            word = word.rstrip('.,:!?)’”')# 末尾のピリオド等を取り除く
            word = word.lstrip('(‘“')     # 冒頭のクォートを取り除く

            if word in freq:                # ③辞書に単語と同じキーがあるか
                freq[word] += 1             # キーの値に1加算
            else:                           # 該当するキーがなければ
                freq[word] = 1              # 単語をキーにして値を1にする
          
    return freq                             # 頻度表としての辞書を返す
```

```python
# 単語の出現回数をカウントして頻度表を作る関数

def get_frequency(file):
    freq = {}                           # 辞書を用意

    file_data = open(file)              # ファイルをオープン
    for line in file_data:              # ①段落を取り出す
        words = line.split()              # 単語に分割してリストにする(改行は除かれる)

        for word in words:                  # ②リストから単語を取り出す
            word = word.rstrip('.,:!?)’”')# 末尾のピリオド等を取り除く
            word = word.lstrip('(‘“')     # 冒頭のクォートを取り除く

            if word in freq:                # ③辞書に単語と同じキーがあるか
                freq[word] += 1             # キーの値に1加算
            else:                           # 該当するキーがなければ
                freq[word] = 1              # 単語をキーにして値を1にする
          
    return freq                             # 頻度表としての辞書を返す
```

```python
# テキストファイルから単語の出現回数順に並べた頻度表を作る
fname = 'alice.txt'         # 読み込むファイルを指定
freq = get_frequency(fname) # ファイルを指定して頻度表の辞書を取得

for word in sorted(
    freq,                   # 対象の辞書
    key=freq.get,           # 並べ替えの基準(key)を辞書の値にする
    reverse=True            # 降順で並べ替え
    ):                        
    print(
      word + '(' + str(freq[word]) + ')' # キー(単語)と値(頻度)を出力
      )
```

```python
# テキストファイルをUTF-8で別ファイルとして保存する関数

def convert_encode(file, encord):
    file_input = open(file + '.txt',        # ファイルをオープン
                      'r',                  # 読み取り専用で開く
                      encoding = encord     # パラメーターで取得したエンコード方式
                      )
    
    file_output = open(file + '_utf-8.txt', # ファイルをオープン(新規作成)
                      'w',                  # 書き換えモードで開く
                      encoding = 'utf-8'    # utf-8で保存する
                      )

    for line in file_input:                 # 改行までを1つの段落として取り出す
        print(line,                         # 段落を書き込む
            file = file_output,             # 出力先をfile_outpuにする
            end  = ''                       # print()独自の改行は出力しない
            )

    file_input.close()                      # Fileオブジェクトをクローズ
```

```python
# テキストファイルをUTF-8で別名で保存
fname = 'alice'     # 変換するファイル名
encoding  = 'cp932' # 変換するファイルのエンコード方式
# 指定したファイルをUTF-8に変換して別名のファイルに保存する
convert_encode(fname, encoding)
```
