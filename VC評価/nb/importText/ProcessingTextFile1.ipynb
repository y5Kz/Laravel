```python
# テキストファイルの中身を出力する

file = open('sample.txt') # ファイルをオープン
for line in file:         # 改行を含めてセンテンス単位で取り出す
  print(line, end='')     # print()自身の改行は行わないで出力
file.close()              # Fileオブジェクトを閉じる
```

```python
# テキストファイルから末尾の改行を取り除き、センテンス単位で出力する

file = open('sample.txt') # ファイルをオープン
for line in file:         # 改行を含めてセンテンス単位で取り出す
  line = line.rstrip()    # 末尾の改行を取り除く
  print(line, end='')     # 出力
file.close()              # Fileオブジェクトを閉じる
```

```python
#テキストファイルの空白行をスキップして出力

file = open('sample_blank.txt') # ファイルをオープン
for line in file:         # 改行を含めてセンテンス単位で取り出す
  line = line.rstrip()    # 末尾の改行を取り除く
  if not line == '':      # 空白行ではない場合に出力
      print(line)
file.close()              # Fileオブジェクトを閉じる
```

```python
# continueを使って空白行をスキップする

file = open('sample_blank.txt')
for line in file:
  line = line.rstrip()
  if line == '': # 空白行の場合
    continue     # スキップして次の繰り返しに進む
  print(line)    # 空白ではないセンテンスだけを出力
file.close()     # Fileオブジェクトを閉じる
```

```python
# 指定した文字列を含む文だけを出力する

file = open('sample.txt')     # ファイルをオープン
for line in file:             # センテンス単位で取り出す
  line = line.rstrip()        # 末尾の改行を取り除く
  if 'アスリート' in line:  # センテンスの中に指定した文字列が含まれるか
    print('見つかりました!')
    print(line)               # 文字列を含むセンテンスを出力
file.close()                  # Fileオブジェクトを閉じる
```

```python
# 特定の文字列を含むセンテンスが見付かった時点でループを止める

file = open('sample.txt')     # ファイルをオープン
for line in file:             # センテンス単位で取り出す
  line = line.rstrip()        # 行末尾の改行を取り除く
  if 'アスリート' in line:  # センテンスの中に指定した文字列が含まれるか
    print('見つかりました!')
    print(line)               # 文字列を含センテンスを出力
    break                     # ループを抜ける
file.close()                  # Fileオブジェクトを閉じる
```

```python
# 段落ごとに連番をふってから出力する

file = open('kusamakura.txt') # ファイルをオープン
counter = 0                   # カウンター変数を初期化
for line in file:             # センテンス単位で取り出す
  counter += 1                # カウンター変数の値を1増やす
  line = line.strip()         # 前後の空白と末尾の改行を取り除く
  print(str(
    counter) + ' : ' + line)  # 番号を付けて段落を出力
file.close()                  # Fileオブジェクトを閉じる
```

```python
# 指定した段落までを表示する関数

def paragraph(fname, pnum):
    file = open(fname)        # ファイルをオープン
    counter = 0               # カウンター変数を初期化
    for line in file:         # 段落単位で取り出す
        line = line.rstrip()  # 末尾の改行のみを取り除く
        print(line)           # 段落を出力
        counter += 1          # カウンター変数の値を1増やす
        if counter == pnum:   # 指定した段落数に達したらループを抜ける
            break
    file.close()              # Fileオブジェクトを閉じる
```

```python
fname = 'kusamakura.txt'      # ファイル名
pnum = 3                      # 段落番号
paragraph(fname, pnum)        # ファイル名と段落番号を指定してparagraph()を実行
```

```python
# 特定の段落のみを出力する関数

def dispParagraph(fname, pnum):
    file = open(fname)       # ファイルをオープン
    counter = 0              # カウンター変数を初期化
    for line in file:        # 段落単位で取り出す
        line = line.rstrip() # 末尾の改行のみを取り除く
        counter += 1         # カウンター変数の値を1増やす
        if counter == pnum:  # 指定した段落数に達したらループを抜ける
            print(line)      # ①段落を出力
            break
    file.close()             # Fileオブジェクトを閉じる
```

```python
fname = 'kusamakura.txt'      # ファイル名
pnum = 3                      # 段落番号
dispParagraph(fname, pnum)    # ファイル名と段落番号を指定してparagraph()を実行
```
